3. Create a function that takes a string and returns a "rotated" version where the first character is moved to the end. For example, "Python" becomes "ythonP".

```python
def rotate_string(s):
    [first, *rest] = s
    return ''.join(rest)  + first

print(rotate_string('Python'))
```

3. Create a function that efficiently concatenates a large number of strings. Compare your approach with using the + operator in a loop.

```python
import time
def efficient_concat(strings):
    return ''.join(strings)

data = ["Python"] * 100_000

start_time = time.time()
print(efficient_concat(data))
print("Time:", round(time.time() - start_time, 4), "seconds") 
```

2. Write a function that counts the occurrences of each character in a string and returns a dictionary with the results.

```python
def counts_occurrences(strings):
    count = {}
    for s in strings:
        count[s] = count.get(s, 0) + 1
        
    return count

print(counts_occurrences('this is python learning'))
```

3. Create a function that checks if a string is a palindrome (reads the same backward as forward), ignoring case, spaces, and punctuation.

```python
def is_palindrome(s: str) -> bool:
    low = 0
    high = len(s) - 1
    
    while low < high:
        if s[low].lower() != s[high].lower():
            return False
        low += 1
        high -= 1
    return True
```

```python
is_palindrome("Python")
```

```python
is_palindrome('Level')
```

4. Create a function that takes a multi-line string and returns the line with the most characters.

```python
def find_most_characters(strings):
    s = strings.split(',\n')
    
    max = len(s[0])
    index = 0
    
    for w in range(0, len(s)):
        if len(s[w]) > max:
            index = w    
    
    return s[index]
    
s = """The woods are lovely, dark and deep,
But I have promises to keep,
And miles to go before I sleep,
And miles to go before I sleep."""
find_most_characters(s)
```

### Section 7: String Formatting (Advanced)

1. Format the same data using all three main formatting methods (f-strings, .format(), and % formatting):
    - Format a floating-point number to two decimal places
    - Format an integer with leading zeros
    - Format a string with fixed width and alignment

```python
# Format a floating-point number to two decimal places
a = 12345.239876379
print(f'a: {a:.2f}')
print('a: {:.2f}'.format(a))
print('a: %.2f' % a)
```

```python
# Format an integer with leading zeros
num = 45
print(f'num: {num:05}')
print('num: {:05}'.format(num))
print('num: %05d' % num)
```

```python
# Format a string with fixed width and alignment
first_name = 'John'
last_name = 'Doe'

print(f'Name: {first_name:<10} {last_name}')
print('Name: {:<10} {}'.format(first_name, last_name))
print('Name: %-10s %s' % (first_name, last_name))
```
